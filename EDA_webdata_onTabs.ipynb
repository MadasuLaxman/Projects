```python
off = []
discount = []
Price =[]
Brands =[]
models =[]
colors =[]
RAM =[]
ROM =[]
display =[]
battery = []
os =[]
warranty =[]
Processor =[]
front_camera =[]
camera =[]
Ratings =[]
Reviews =[]
ratings = []
for i in range(9):
    print("""https://www.flipkart.com/search?q=tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i))
    url = """https://www.flipkart.com/search?q=tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i)
    page = requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    headers = {"Accept-Language": "en-US,en;q=0.9"}
    page=requests.get(url, headers=headers)
    
    
    soup =BeautifulSoup(page.text)
    
    a = soup.find_all('div', class_='_2kHMtA')
    
    for i in a:
        bat = i.find("div",class_="_30jeq3 _1_WHN1").text
        regex = re.findall("\S\d+,\d+",bat)[0]
        if regex:
            Price.append(regex)
        else:
            Price.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        Brands.append(i.text.split()[0])
        
    for i in soup.find_all("div",class_="_3Ay6Sb"):
        discount.append(i.text)
    else:
        discount.append(np.nan)   
        
    for i in soup.find_all("div",class_="_3I9_wc _27UcVY"):
        off.append(i.text)
    else:
        off.append(np.nan)
        
    for i in soup.find_all("div",class_="_4rR01T"):
        colors.append(i.text.split("(")[-1].replace(")",""))
    
    for i in a:
        bat = i.find("div",class_="gUuXy-").text
        regex = re.findall("\d+.\d(.*)\sRatings",bat)
        if regex:
            ratings.append(regex[0])
        else:
            ratings.append(np.nan)
    
    for i in a:
        bat = i.find("span",class_="_2_R_DZ").text
        regex = re.findall("\&\s(.*)\sReviews",bat)
        if regex:
            Reviews.append(regex[0])

        else:
            Reviews.append(np.nan)
            
    for i in a:
        bat = i.find("div",class_="gUuXy-").text
        regex = re.findall("(\d.\d)",bat)
        if regex:
            Ratings.append(regex[0])
        else:
            Ratings.append(np.nan)
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\s\w+\sRAM",bat)
        if regex:
            RAM.append(regex[0])
        else:
            RAM.append(np.nan)
            
    for i in  a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\s\w+\sROM",bat)
        if regex:
            ROM.append(regex[0])
        else:
            ROM.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+\.\d+)\s\wnch",bat)
        if regex:
            display.append(regex[0])
        else:
            display.append(np.nan)
     
    
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\sMP\s\w+\sCamera",bat)
        if regex:
            camera.append(regex[0])
        else:
            camera.append(np.nan)
            
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\sMP\sFront",bat)
        if regex:
            front_camera.append(regex[0])
        else:
            front_camera.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall('(\d+)\smAh', bat)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("Processor:\s(\w+\s\w+)",bat)
        if regex:
            Processor.append(regex[0])
        else:
            Processor.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d)\sYear",bat)
        if regex:
            warranty.append(regex[0])
        else:
            warranty.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("Front(\w+\s\d{2}\s)",bat)
        if regex:
            os.append(regex[0])
        else:
            os.append(np.nan)
            
    f=soup.find_all("div",class_="_4rR01T")
    
    for i in f:
        a= i.text
        b = re.findall("[a-zA-Z]+\s\w+\s\w+",a)
        if len(b)>0:
            models.append(b[0])
for i in range(19,30):
    print("""https://www.flipkart.com/search?q=tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i))
    url = """https://www.flipkart.com/search?q=tablets&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="""+str(i)
    page = requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    headers = {"Accept-Language": "en-US,en;q=0.9"}
    page=requests.get(url, headers=headers)
    
    
    soup =BeautifulSoup(page.text)
    
    a = soup.find_all('div', class_='_2kHMtA')
    
    for i in a:
        bat = i.find("div",class_="_30jeq3 _1_WHN1").text
        regex = re.findall("\S\d+,\d+",bat)[0]
        if regex:
            Price.append(regex)
        else:
            Price.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        Brands.append(i.text.split()[0])
        
    for i in soup.find_all("div",class_="_3Ay6Sb"):
        discount.append(i.text)
    else:
        discount.append(np.nan)   
        
    for i in soup.find_all("div",class_="_3I9_wc _27UcVY"):
        off.append(i.text)
    else:
        off.append(np.nan)
        
    for i in soup.find_all("div",class_="_4rR01T"):
        colors.append(i.text.split("(")[-1].replace(")",""))
    
    for i in a:
        bat = i.find("div",class_="gUuXy-").text
        regex = re.findall("\d+.\d(.*)\sRatings",bat)
        if regex:
            ratings.append(regex[0])
        else:
            ratings.append(np.nan)
    
    for i in a:
        bat = i.find("span",class_="_2_R_DZ").text
        regex = re.findall("\&\s(.*)\sReviews",bat)
        if regex:
            Reviews.append(regex[0])

        else:
            Reviews.append(np.nan)
            
    for i in a:
        bat = i.find("div",class_="gUuXy-").text
        regex = re.findall("(\d.\d)",bat)
        if regex:
            Ratings.append(regex[0])
        else:
            Ratings.append(np.nan)
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\s\w+\sRAM",bat)
        if regex:
            RAM.append(regex[0])
        else:
            RAM.append(np.nan)
            
    for i in  a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\s\w+\sROM",bat)
        if regex:
            ROM.append(regex[0])
        else:
            ROM.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+\.\d+)\s\wnch",bat)
        if regex:
            display.append(regex[0])
        else:
            display.append(np.nan)
     
    
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\sMP\s\w+\sCamera",bat)
        if regex:
            camera.append(regex[0])
        else:
            camera.append(np.nan)
            
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d+)\sMP\sFront",bat)
        if regex:
            front_camera.append(regex[0])
        else:
            front_camera.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall('(\d+)\smAh', bat)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("Processor:\s(\w+\s\w+)",bat)
        if regex:
            Processor.append(regex[0])
        else:
            Processor.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("(\d)\sYear",bat)
        if regex:
            warranty.append(regex[0])
        else:
            warranty.append(np.nan)
            
    for i in a:
        bat = i.find("ul",class_="_1xgFaf").text
        regex = re.findall("Front(\w+\s\d{2}\s)",bat)
        if regex:
            os.append(regex[0])
        else:
            os.append(np.nan)
            
    f=soup.find_all("div",class_="_4rR01T")
    
    for i in f:
        a= i.text
        b = re.findall("[a-zA-Z]+\s\w+\s\w+",a)
        if len(b)>0:
            models.append(b[0])
```
